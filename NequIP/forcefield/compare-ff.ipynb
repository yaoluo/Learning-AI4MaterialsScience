```python
import numpy as np 
import matplotlib.pyplot as plt 

plt.figure(figsize=(6,6))
predicted_f_abs =np.load("twobody-ff/|F|_nn.npy")
val_f_abs = np.load("twobody-ff/|F|_validation.npy")
plt.scatter(val_f_abs, predicted_f_abs,s=10,alpha=0.5,zorder=-1,label='Twobody-NN')


predicted_f_abs =np.load("nequ-ff/|F|_nn.npy")
val_f_abs = np.load("nequ-ff/|F|_validation.npy")
plt.scatter(val_f_abs, predicted_f_abs,s=10,alpha=0.5,zorder=-1,label='Equivariant-GNN')

plt.legend(fontsize=15)
plt.xlim([0,150])
plt.ylim([0,150])
plt.plot([0,150],[0,150],zorder=1,c = 'k', linestyle='--')
plt.xlabel(r'|Force|$_{QM}$', fontsize=20)
plt.ylabel(r'|Force|$_{ML}$', fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.tight_layout()

plt.savefig("compare-forcefield.jpg", dpi=500)
```
